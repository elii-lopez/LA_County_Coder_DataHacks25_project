### **Cleaning up the TMBD dataset**

The TMBD dataset containts columns and data that we will not need for our project. In this part of the project we aim to clean up our dataset for readability and ease of use purposes

````python
# prompt: make the index of reveunie df the 'original_title' column, remove any movie that 'status' != released, remove 'backdrop column' & "backdrop path","imdb_id","overveiw", "poster_path","tagline". make sure to assign the index in a diff df so i can keep my original df intact

revenue_df_indexed = revenue_df[revenue_df['status'] == 'Released'].copy()
revenue_df_indexed = revenue_df_indexed.set_index('original_title')
revenue_df_indexed = revenue_df_indexed.drop(columns=['backdrop_path', 'imdb_id', 'overview', 'poster_path', 'tagline','homepage','keywords','production_countries','spoken_languages','original_language','id'])

````

````python
# prompt: in my revenue_df_indexed for release date is formatted YYYY-MM-DD keep only YYYY. Example of data input: 2010-07-15. I would only want to keep 2010

revenue_df_indexed['release_year'] = pd.to_datetime(revenue_df_indexed['release_date']).dt.year
revenue_df_indexed = revenue_df_indexed.drop(columns=['release_date','title'])
````

````python
# prompt: get the revenue_df_indexed and make a new column called 'profit' which gets the 'revenue' - 'budget' . Then sort by vote count increasing

#revenue_df_indexed['profit'] = revenue_df_indexed['revenue'] - revenue_df_indexed['budget']
#revenue_df_indexed = revenue_df_indexed.sort_values('vote_count')
#revenue_df_indexed.head()

````

````python
# Convert revenue and budget to numeric (if not already)
revenue_df_indexed['revenue'] = pd.to_numeric(revenue_df_indexed['revenue'], errors='coerce')
revenue_df_indexed['budget'] = pd.to_numeric(revenue_df_indexed['budget'], errors='coerce')
````

````python
# Recalculate profit
revenue_df_indexed['profit'] = revenue_df_indexed['revenue'] - revenue_df_indexed['budget']

# Remove movies with extreme losses (>$263.7M)
revenue_df_indexed = revenue_df_indexed[
    (revenue_df_indexed['vote_count'] > 100) &
    (revenue_df_indexed['revenue'] > 0) &
    (revenue_df_indexed['budget'] > 0) &
    (revenue_df_indexed['revenue'].notna()) &
    (revenue_df_indexed['budget'].notna()) &
    (revenue_df_indexed['profit'] > -263_700_000)  # changed this!
]

# Sort to confirm
revenue_df_indexed = revenue_df_indexed.sort_values('profit')
revenue_df_indexed.head()



````

### **Cleaning up the "Movies" dataset**

The "Movies" dataset containts columns and data that we will not need for our project. In this part of the project we aim to clean up our dataset for readability and ease of use purposes


````python
budget_df = pd.read_csv('movie_dataset.csv.zip')
````

````python
budget_df_indexed = budget_df[budget_df['status'] == 'Released'].copy()
budget_df_indexed = budget_df_indexed.set_index('original_title')
budget_df_indexed = budget_df_indexed.drop(columns=['index', 'homepage', 'id','tagline','title','crew','runtime','production_countries','keywords','status','keywords','overview'])
````

````python
len(budget_df_indexed)
````

````python
# prompt: remove any movies with column 'original_language' = 'en'

# Assuming 'original_language' was not dropped earlier
budget_df_indexed = budget_df_indexed[budget_df_indexed['original_language'] == 'en']
budget_df_indexed = budget_df_indexed.drop(columns=['spoken_languages','original_language','release_date','vote_average'])
budget_df_indexed.head()
````

````python
budget_df_indexed = budget_df_indexed.reset_index()
````

````python
revenue_df_indexed = revenue_df_indexed.reset_index()
````

````python
# prompt: make all titles in the original title column lower case in the remove special character from movie titles. exampe: ; , ' " ? + -

import re

# Function to remove special characters from movie titles
def remove_special_characters(title):
    title = re.sub(r'[^\w\s]', '', title) # Remove special characters
    return title

# Apply the function to the 'original_title' column and convert to lowercase
revenue_df_indexed['original_title'] = revenue_df_indexed['original_title'].apply(lambda x: remove_special_characters(str(x).lower()))
budget_df_indexed['original_title'] = budget_df_indexed['original_title'].apply(lambda x: remove_special_characters(str(x).lower()))

````

### **Merging the two new datasets with cleaned up entries**

merging on title names to keep only popular movies.


````python
budget_df_indexed.head()
````

````python
revenue_df_indexed.head()
````

````python
# prompt: lower case all the movie titles in the revenue df

revenue_df_indexed['original_title'] = revenue_df_indexed['original_title'].str.lower()

````

````python
# prompt: lower case all the movie titles in the budget_df_indexdf

budget_df_indexed['original_title'] = budget_df_indexed['original_title'].str.lower()

````

````python
revenue_df_indexed['original_title'] = (
    revenue_df_indexed['original_title']
    .str.lower()
    .str.replace(' ', '', regex=False)  # removes *all* spaces
)

revenue_df_indexed.head()
````

````python
budget_df_indexed['original_title'] = (
    budget_df_indexed['original_title']
    .str.lower()
    .str.replace(' ', '', regex=False)  # removes all spaces from the oringal_title column
)
budget_df_indexed.head()
````

````python
merged_movies_df = pd.merge(
    budget_df_indexed,
    revenue_df_indexed,
    on='original_title',
    how='inner'  # only keeps rows where titles match in both
)

# Preview the merged result
merged_movies_df.head()
````
